```python
import matplotlib.pyplot as plt
from mpl_flow import Flow
fig, axes = plt.subplots(1, 1, figsize=(5, 6))
ax1 = axes
f = Flow(ax=ax1)

f.node('s', label=" Student ", bbox=dict(boxstyle='circle'), connect=False)
f.node('l', label=" Lecturer ", bbox=dict(boxstyle='circle'), travel='n')
f.node('t', label=" Tutor ", bbox=dict(boxstyle='circle'), connect=False)
f.node('m', label=" Mentor ", bbox=dict(boxstyle='circle'), startpoint='s', travel='e')
f.node('h', label=" Head of\nProgram ", bbox=dict(boxstyle='circle'), startpoint='l', travel='n')
f.node('p', label=" PIC ", bbox=dict(boxstyle='circle'), startpoint='t', travel='n')
f.node('v', label=" Vice\nDean ", bbox=dict(boxstyle='circle'), startpoint='h', travel='n')
f.edge('l', 'p', tailport='e', headport='w')
f.edge('s', 't', tailport='e', headport='w', arrowprops=dict(connectionstyle='arc3,rad=-0.2'))
f.edge('m', 'p', tailport='e', headport='e', arrowprops=dict(connectionstyle='arc3,rad=-0.8'))
plt.show()
```
