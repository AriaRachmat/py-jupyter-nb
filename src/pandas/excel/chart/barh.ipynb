# barh
Read excel file using pandas and plot horizontal bar chart using matplotlib

## read dataframe

```python
import pandas as pd
df = pd.read_excel('data/bar.xlsx', sheet_name='PYPL2022')
lang = df['Language'].tolist()
print(lang)
```

## plot for january 2022

```python
jan = [i*100 for i in df['Jan'].tolist()]

import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.rc('font', size=20)
plt.xlabel("Popularity (%)")
plt.ylabel("Programming languages")
plt.barh(lang, jan, color='#aaaaff', height=0.9)
plt.legend(['Jan 2022'])
plt.show()
```

## plot for february 2022

```python
feb = [i*100 for i in df['Feb'].tolist()]

import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.rc('font', size=20)
plt.figure(figsize=(10,10))
plt.rc('font', size=20)
plt.xlabel("Popularity (%)")
plt.ylabel("Programming languages")
plt.barh(lang, jan, color='tab:blue', height=0.9)
plt.legend(['Feb 2022'])
plt.show()
```

## multiple bars chart (different width approach)

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
plt.rc('font', size=20)
plt.xlabel("Popularity (%)")
plt.ylabel("Programming languages")
plt.barh(lang, jan, color='#aaaaff', height=0.9)
plt.barh(lang, feb, color='tab:blue', height=0.4)
plt.legend(['Jan 2022', 'Feb 2022'])
plt.show()
```
