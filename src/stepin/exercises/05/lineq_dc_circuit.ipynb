# lineq_dc_circuit
+ System of linear equations for DC circuit.

```python
import schemdraw
import schemdraw.elements as elm
with schemdraw.Drawing(file='schematic.svg') as d:
    d += elm.Resistor().label('100KΩ')
    d += elm.Capacitor().down().label('0.1μF', loc='bottom')
    d += elm.Line().left()
    d += elm.Ground()
    d += elm.SourceV().up().label('10V')
```

```python
import schemdraw
import schemdraw.elements as elm
with schemdraw.Drawing(file='schematic.svg') as d:
    d += elm.Resistor().label('$R_2$')
    d += elm.BatteryCell().down().label('$\epsilon_2$', loc='bottom')
    d += elm.Resistor().left().label('$R_4$', loc='bottom')
    d += elm.Resistor().up().label('$R_5$')
    d += elm.Dot()
    d += elm.Resistor().left().label('$R_1$')
    d += elm.BatteryCell().down().label('$\epsilon_1$', loc='top')
    d += elm.Dot()
    d.push()
    d += elm.Resistor().right().label('$R_3$', loc='bottom')
    d.pop()
    d += elm.BatteryCell().down().label('$\epsilon_3$', loc='top')
    d += elm.Resistor().right().label('$R_6$', loc='top')
    d += elm.Dot()
    d.push()
    d += elm.Resistor().right().label('$R_8$', loc='top')
    d += elm.BatteryCell().up().reverse().label('$\epsilon_4$', loc='bottom')
    d += elm.Dot()
    d.pop()
    d += elm.Resistor().up().label('$R_7$', loc='top')
```
